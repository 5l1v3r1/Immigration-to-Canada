# Pie Chats

```python
# group countries by continents and apply sum() function 
df_continents = datasets.groupby('Continent', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(datasets.groupby('Continent', axis=0)))

df_continents.head()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind= 'pie', figsize=(15.5, 7), startangle=90, autopct='%1.1f%%',
                    shadow=True, labels=None, pctdistance=1.10,colors=colors_list, explode=explode_list)

plt.title('Immigrations to Canada by Continent from 1980 - 2013', y= 1.10)

plt.legend(labels= df_continents.index, loc= 'upper left')
plt.axis('equal')

plt.show()
```

### We want to Create a box plot to visualize the distribution of the top 15 countries (based on total immigration) grouped by the decades 1980s, 1990s, and 2000s.

```python
# We get the dataset. Get the top 15 countries based on Total immigrant population. Name the dataframe df_top15
df_top15 = datasets.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

```python
# First Creat all Years in decades in 80s, 90, 00s.
years_80s = list(map(str, range(1980, 1990)))
years_90s = list(map(str, range(1990, 2000)))
years_00s = list(map(str, range(2000, 2010)))

# Slice the original Dataframe datasets to create a series for each decade
df_80s = df_top15.loc[:, years_80s].sum(axis=1)
df_90s = df_top15.loc[:, years_90s].sum(axis=1)
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

# we merge the three series in to a new dataframe

new_df = pd.DataFrame({'years_80s':df_80s, 'years_90s': df_90s, 'years_00s':df_00s})

new_df.head(15)
```

```python
new_df.describe()
```

```python
new_df.plot(kind = 'box', figsize=(9, 5))

plt.title('Immigration of top 15 Countries to Canada from 1980s - 2000s')

plt.show()
```

## Scatter Plots

```python
df_tol = pd.DataFrame(datasets[years].sum(axis=0))


# We change the years to type int (useful for regression later on)
df_tol.index = map(int, df_tol.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tol.reset_index(inplace=True)

# rename columns
df_tol.columns = ['year', 'total']

df_tol.head()
```

```python
df_tol.plot(kind='scatter', x='year', y = 'total', figsize=(10, 8), color ='blue')

plt.title('Total immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')


plt.show()
```

**Notice how the scatter plot does not connect the datapoints together. We can clearly observe an upward trend in the data: as the years go by, the total number of immigrants increases. We can mathematically analyze this upward trend using a regression line (line of best fit)**

```python
# Year on x axis
x = df_tol['year']
# Total on y axis
y = df_tol['total']

fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tol.plot(kind='scatter', x='year', y = 'total', figsize=(10, 8), color ='blue')

plt.title('Total immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# Plot the line of best fit
plt.plot(x, fit[0] * x + fit[1], color='blue')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```
